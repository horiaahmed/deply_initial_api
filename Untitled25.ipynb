```python
import numpy as np
import cv2
import mediapipe as mp
import time
import pandas as pd

mp_drawing = mp.solutions.drawing_utils
mp_hands = mp.solutions.hands
hands = mp_hands.Hands()

cap = cv2.VideoCapture(r'E:\Downloads\Telegram Desktop\video_2024-11-08_21-34-56.mp4')
cap.set(3, 1280)
cap.set(4, 720)
pTime = 0
Hand_landmarks = []  # Initialize Hand_landmarks as an empty list

while True:
    ret, image = cap.read()
    if not ret:
        print("Failed to read frame")  # Debug statement
        break

    # Process the image for hand landmarks
    image = cv2.cvtColor(cv2.flip(image, 1), cv2.COLOR_BGR2RGB)
    results = hands.process(image)
    rows, cols, _ = image.shape
    print("Image dimensions:", rows, cols)  # Debug statement

    # Check if hand landmarks are detected
    if results.multi_hand_landmarks:
        print("Hand landmarks detected")  # Debug statement
        for hand_landmarks in results.multi_hand_landmarks:
            mp_drawing.draw_landmarks(image, hand_landmarks, mp_hands.HAND_CONNECTIONS)
            
            # Extract and store landmark positions
            hand_data = []
            for lm in hand_landmarks.landmark:
                cx, cy = int(lm.x * cols), int(lm.y * rows)
                hand_data.append((cx, cy))
                cv2.circle(image, (cx, cy), 5, (0, 0, 255), -1)
            Hand_landmarks.append(hand_data)  # Append hand landmark data
            print("Added hand data:", hand_data)  # Debug statement
    else:
        print("No hand landmarks detected")  # Debug statement

    image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)
    cTime = time.time()
    fps = 1 / (cTime - pTime)
    pTime = cTime
    cv2.putText(image, f'FPS: {int(fps)}', (70, 50), cv2.FONT_HERSHEY_PLAIN, 3, (0, 0, 255), 3)
    cv2.imshow('MediaPipe Hands', image)

    if cv2.waitKey(1) & 0xFF == 27:  # ESC key to exit
        break

cap.release()
cv2.destroyAllWindows()

# Print the collected hand landmark data
print("Collected hand landmark data:", Hand_landmarks)
```

```python
import pandas as pd

# Convert Hand_landmarks data to a DataFrame
Hand_landmarks_df = pd.DataFrame(Hand_landmarks)

# Display the first few rows and the shape of the DataFrame
print(Hand_landmarks_df.head())
print(Hand_landmarks_df.shape)

# Add a 'label' column with a value of 25
Hand_landmarks_df['label'] = 18

# Display the first few rows and the shape of the DataFrame again
print(Hand_landmarks_df.head())
print(Hand_landmarks_df.shape)

# Save the DataFrame to a CSV file
Hand_landmarks_df.to_csv('s.csv', index=False)
```
